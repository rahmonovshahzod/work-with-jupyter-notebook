question 1: Use yfinance to Extract Stock Data for Tesla

```python
import yfinance as yf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Download Tesla's stock data
tesla_data = yf.download('TSLA', start='2020-01-01', end='2022-12-31')

# Reset the index
tesla_data.reset_index(inplace=True)

# Display the first five rows
print(tesla_data.head())
```

question 2: Use Webscraping to Extract Tesla Revenue Data

```python
import requests
from bs4 import BeautifulSoup

# URL of the page to be scraped
url = "https://www.macrotrends.net/stocks/charts/TSLA/tesla/revenue"

# Send a GET request to the webpage
response = requests.get(url)

# Parse the webpage
soup = BeautifulSoup(response.text, 'html.parser')

# Find the table with the revenue data
table = soup.find_all('table')[1]

# Extract the column names and data from the table
column_names = [item.get_text() for item in table.find_all('th')]
data = [[item.get_text() for item in row.find_all('td')] for row in table.find_all('tr')]

# Create a DataFrame
tesla_revenue = pd.DataFrame(data, columns=column_names)

# Display the last five rows
print(tesla_revenue.tail())
```

5: Plot Tesla Stock Graph

```python
import matplotlib.pyplot as plt

def make_graph(stock_data, title):
    plt.figure(figsize=(10,6))
    plt.plot(stock_data['Date'], stock_data['Close'])
    plt.title(title)
    plt.xlabel('Date')
    plt.ylabel('Close Price')
    plt.show()

# Call the function with the desired data
make_graph(tesla_data, 'Tesla Stock Data')
```
